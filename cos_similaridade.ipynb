```python
# --- Sistema de Recomendação por Similaridade de Cosseno: 
# O modelo utiliza uma matriz para verificar a congruência em que dois produtos são vendidos na mesma compra ---
# Objetivo: Recomendação de produto para clientes novos.
# O arquivo segue uma organização de arquivos /cos_similaridade.ipynb
# /datasets/vendas_2023_2024.csv

# Importação das bibliotecas pandas para manipular dados e sklearn para construção da matriz
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Leitura do .csv de vendas
df = pd.read_csv('datasets/vendas_2023_2024.csv')
```

```python
# Construção e Exibição da matriz comparando clientes e produtos
matriz = pd.crosstab(df['id_client'], df['id_product'])
matriz
```

```python
# Alteração de valores quantitativos para True ou False, usando 1 e 0
matriz_presenca = (matriz > 0).astype(int)
matriz_presenca
```

```python
# Fórmula para cálculo de similaridade na matriz
similarity = cosine_similarity(matriz_presenca.T)
```

```python
# Formatação e Exibição dos valores de similaridades
df_similarity = pd.DataFrame(
    similarity, 
    index=matriz_presenca.columns, 
    columns=matriz_presenca.columns
)
df_similarity
```

```python
# Determinação para produto alvo
produto_alvo = 27
# Atribuição para organizar os valores de similaridade em função do produto alvo
similares = df_similarity[produto_alvo].sort_values(ascending=False)
```

```python
# Leitura da tabela de produtos tratados e renomeação da coluna 'code' para 'id_product'
df_produtos = pd.read_csv('produtos_raw_tratados.csv')
df_produtos.rename(columns={'code':'id_product'}, inplace=True)
```

```python
# Junção e Exibição das tabelas com valor de similaridade e produtos
df_final = pd.merge(similares, df_produtos, on='id_product')
df_final.rename(columns={27:'similaridade'},inplace=True)
df_final
```
