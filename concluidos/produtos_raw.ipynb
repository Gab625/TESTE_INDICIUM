```python
import pandas as pd
import numpy as np

df = pd.read_csv("../datasets/produtos_raw.csv")
df
```

```python
df['price'] = df['price'].str.replace('R$', '',regex=False).astype('float64')
df['price']
```

```python
df['actual_category'] = df['actual_category'].str.lower().str.replace(' ','')
df['actual_category']
```

```python
condicoes = [
    df['actual_category'].str.contains('eletr'),
    df['actual_category'].str.contains('pro'),
    df['actual_category'].str.contains('anco|enc'),
]

normalizada = ['eletrônicos', 'propulsão', 'ancoragem']
```

```python
df['coluna_normalizada'] = np.select(condicoes, normalizada, default='Outros')
print(df['coluna_normalizada'].value_counts())
```
